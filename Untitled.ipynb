```python
# Handle outliers
q1 = df['Rotational speed [rpm]'].quantile(0.25)
q3 = df['Rotational speed [rpm]'].quantile(0.75)
iqr = q3 - q1
upper_bound = q3 + 1.5 * iqr
df = df[df['Rotational speed [rpm]'] <= upper_bound]
```

```python
df.boxplot(figsize=(10,5))
```

### Identify redundance rows

```python
# Analysing redundance value, count failures on "Target" column
count_failures = (df['Target'] == 1).sum()
print('Number of failures:', count_failures)
# Sum of occurrences of 'Failure_type' other than 'No Failure'
sum_failures = df.loc[(df['Failure Type'] != 'No Failure') & (df['Target'] == 1), 'Failure Type'].count()
print('Sum of failures different of "No Failure":', sum_failures)
```

```python
# Filter the rows with has 1 in the column "Target" and 'No Failure' in the column 'Failure_type'
rows_to_remove = df[(df['Target'] == 1) & (df['Failure Type'] == 'No Failure')]
# Remove this filtered row in the main df
df.drop(rows_to_remove.index)
```

```python
# Removing "Random Failures" of the "Failure_type" column.
df = df.loc[df['Failure Type'] != 'Random Failures']
```

```python
df['Failure Type'].unique()
```

### EDA

```python
#correlation matrix to see relationships btw features and the target
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix,annot=True, cmap='coolwarm')
plt.show()
```

```python
# Distribution of numerical features
numerical_features = ['Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']
for feature in numerical_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[feature], kde=True)
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Frequency")
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="Type", y="Tool wear [min]", data=df, palette="pastel")
plt.title("Box Plot of Tool Wear by Product Type")
plt.show()
```

### Failure Rate Distribution of columns with respect to Failure Status

```python
# 1. Failure Rate Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Target', palette='viridis')
plt.title('Failure Rate Distribution')
plt.xlabel('Failure Status (0 = No Failure, 1 = Failure)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df[df["Target"] == 1]["Process temperature [K]"], bins=30, kde=True, color="crimson")
plt.title("Histogram of Process Temperature for Failures")
plt.xlabel("Process Temperature [K]")
plt.ylabel("Frequency")
plt.show()
```

```python
#  Torque vs Temperature by Failure Status
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Torque [Nm]', y='Process temperature [K]', hue='Failure Type')
plt.title("Torque vs Process Temperature by Failure Type")
plt.xlabel("Torque")
plt.ylabel("Process Temperature")
plt.legend(title="Failure Type")
plt.show()
```

Above scatter plot visualizes the relationship between torque and temperature, colored by failure type. Clusters or trends in failure points indicate that certain torque and temperature combinations are linked to a higher likelihood of failure.

```python
# Count types
tipo_contagem = df['Type'].value_counts()

# Show graphic
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='Type')
plt.title('Proportion of Types')
plt.xlabel('Type')
plt.ylabel('Count')

# Add as percentages as labels
total = len(df['Type'])
for p in ax.patches:
    height = p.get_height()
    percentage = (height / total) * 100
    ax.annotate(f'{percentage:.2f}%', (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom')
plt.show()
```

```python
# Calculate the distribution of failure classes
proportions = df['Target'].value_counts(normalize=True)
proportions.index = ['No Failure', 'Failure']  # Rename index for clarity

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(proportions, labels=proportions.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'salmon'])
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Failure vs. Non-Failure Class Distribution')
plt.show()
```

```python
# Filter to include only failure instances and calculate the distribution of failure types
failure_types = df[df['Target'] == 1]['Failure Type'].value_counts(normalize=True)

# Plot the pie chart for failure types
plt.figure(figsize=(8, 6))
plt.pie(failure_types, labels=failure_types.index, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Distribution of Failure Types')
plt.show()
```

```python
# 2. Failure Probability by Air Temperature Range
# Create temperature categories (e.g., <300K, 300-301K, >301K)
df['Temperature Range'] = pd.cut(df['Air temperature [K]'], bins=[df['Air temperature [K]'].min(), 300, 301, df['Air temperature [K]'].max()], labels=['<300K', '300-301K', '>301K'])

# Calculate failure rates by temperature range
temp_failure_rate = df.groupby('Temperature Range')['Target'].mean() * 100  # Convert to percentage

# Plot failure probability by temperature range
plt.figure(figsize=(8, 6))
sns.barplot(x=temp_failure_rate.index, y=temp_failure_rate.values, palette='viridis')
plt.legend(['<300K: Low Risk', '300-301K: Moderate Risk', '>301K: High Risk'], loc='upper left', title="Temperature Ranges")
plt.title('Failure Probability by Air Temperature Range')
plt.xlabel('Air Temperature Range (K)')
plt.ylabel('Failure Probability (%)')
plt.show()
```

```python
# 1. Failure Rate by Tool Wear Threshold
# Create a new column for tool wear categories (e.g., <=140 minutes and >140 minutes)
df['Tool Wear Category'] = pd.cut(df['Tool wear [min]'], bins=[0, 140, df['Tool wear [min]'].max()], labels=['<=140 min', '>140 min'])

# Calculate failure rates by tool wear category
tool_wear_failure_rate = df.groupby('Tool Wear Category')['Target'].mean() * 100  # Convert to percentage

# Plot failure rates by tool wear category
plt.figure(figsize=(8, 6))
sns.barplot(x=tool_wear_failure_rate.index, y=tool_wear_failure_rate.values, palette='coolwarm')
plt.legend(['<=140 min: Lower Failure Risk', '>140 min: Higher Failure Risk'], loc='upper left', title="Tool Wear Thresholds")
plt.title('Failure Rate by Tool Wear Threshold')
plt.xlabel('Tool Wear Category')
plt.ylabel('Failure Rate (%)')
plt.show()
```
